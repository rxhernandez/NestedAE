```python

# Set golbal random seed
tf.random.set_seed(1)

# Using kFold strategy 

n_splits = 6
size = 4
#random_state = np.random.randint(0,500)
#random_state = 468
#print(random_state)

i = 0

kf = KFold(n_splits=n_splits, shuffle=False)

for train_index, test_index in kf.split(image_dataset):
    
    print(train_index.shape)
    print(test_index.shape)
    
    # Split into images into train and test sets
    train_images_144_144, test_images_144_144 = image_dataset_144_144_shuffled[train_index], image_dataset_144_144_shuffled[test_index]
    train_images_72_72, test_images_72_72 = image_dataset_72_72_shuffled[train_index], image_dataset_72_72_shuffled[test_index]
    train_images_36_36, test_images_36_36 = image_dataset_36_36_shuffled[train_index], image_dataset_36_36_shuffled[test_index]
    train_images_12_12, test_images_12_12 = image_dataset_12_12_shuffled[train_index], image_dataset_12_12_shuffled[test_index]
    
    # Split cont attributes into train and test sets
    skin_attribute_cont_train, skin_attribute_cont_test = skin_attribute_cont_shuffled[train_index], skin_attribute_cont_shuffled[test_index]
    body_attribute_cont_train, body_attribute_cont_test = body_attribute_cont_shuffled[train_index], body_attribute_cont_shuffled[test_index]
    env_attribute_cont_train, env_attribute_cont_test = env_attribute_cont_shuffled[train_index], env_attribute_cont_shuffled[test_index]
    
    epochs = 200
    period = 1
    batch_size = 1
    
    train_images = {
        '144_144':train_images_144_144,
        '36_36':train_images_36_36
    }
    
    test_images = {
        '144_144':test_images_144_144,
        '36_36':test_images_36_36
    }
    
    train_data = tf.data.Dataset.from_tensor_slices((train_images['144_144'],
                                                     (train_images['144_144'],
                                                      train_images['36_36'])))
    
    train_data = train_data.batch(batch_size)
    train_data = train_data.prefetch(buffer_size=tf.data.AUTOTUNE)
    
    test_data = tf.data.Dataset.from_tensor_slices((test_images['144_144'],
                                                     (test_images['144_144'],
                                                      test_images['36_36'])))
    
    test_data = test_data.batch(batch_size)
    test_data = test_data.prefetch(buffer_size=tf.data.AUTOTUNE)
    
    # Choose the model object you want to initialize
    #model=SCA()
    #model=SCA_2()
    model=CVA(10, 16, "bilinear")
    
    # Select the model type
    #model_type='sca'
    #model_type='sca_2'
    model_type='cva'
    
    # Select the loss functions
    loss_list=['mae', 'mae']
    
    # Select the loss weights
    loss_wts=[1,1]
    
    # Select the metrics you want to use
    metrics_list=['mae', 'mae']
    
    model = compile_model(model, model_type, loss_list, loss_wts, metrics_list)
    
    # Specify date in format yyyy-mm-dd or just use today's date
    dir_name = str(date.today()) + "_0"
    print(dir_name)
    
    model_train_hist, model_test_hist, model = fit_eval_model(model, model_type, epochs, 
                                                              batch_size, dir_name, period, 
                                                              train_data, test_data)
    
    if i == 0:
        break
    
    i += 1

print(model_test_hist)
```

### Training Supervised Autoencoder

```python
# Shuffling the features and labels
seed = 600
from sklearn.utils import shuffle

latent_shuffled = shuffle(X_sa, random_state=seed)
labels_shuffled = shuffle(y_sa, random_state=seed)

kf = KFold(n_splits=6, shuffle=False)

i = 0

for train_index, test_index in kf.split(latent_shuffled):
    
    print(train_index.shape)
    print(test_index.shape)
    
    latent_train, latent_test = latent_shuffled[train_index], latent_shuffled[test_index]
    labels_train, labels_test = labels_shuffled[train_index], labels_shuffled[test_index]
    
    print("latent train shape : {}".format(latent_train.shape))
    print("latent test shape : {}".format(latent_test.shape))
    
    print("labels train shape : {}".format(labels_train.shape))
    print("labels test shape : {}".format(labels_test.shape))
    
    # Compiling SA model
    sa_train = SA(latent_train.shape[1], 0, 32)
    sa_train = compile_sa(sa_train)
                
    # Specify date in format yyyy-mm-dd or just use today's date
    dir_name = str(date.today())
    print(dir_name)
                
    train_hist_sa, test_hist_sa, sa_train = fit_eval_sa(sa_train, 50, 1, dir_name, 1,
                                                        latent_train, latent_test,
                                                        labels_train, labels_test)
    
    if i == 0:
        break
```
